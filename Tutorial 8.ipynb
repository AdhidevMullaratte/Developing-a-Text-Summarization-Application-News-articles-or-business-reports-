```python
import re
import torch
import nltk
from transformers import BartForConditionalGeneration, BartTokenizer
```

```python
def clean_text(text):
    """
    Cleans text by removing special characters, extra spaces, and newlines.
    """
    text = re.sub(r"\s+", " ", text)  # Remove extra spaces
    text = re.sub(r"\n", " ", text)  # Remove newlines
    text = re.sub(r"[^\w\s.]", "", text)  # Remove special characters except periods
    return text.strip()

# Example text
text = """Natural language processing (NLP) is a field of AI that focuses on the interaction between
          computers and humans using natural language. The goal is to enable computers to understand,
          interpret, and generate human language in a meaningful way."""

cleaned_text = clean_text(text)
print("Cleaned Text:", cleaned_text)
```

```python
import torch
from datasets import load_dataset
from transformers import BartForConditionalGeneration, BartTokenizer, Seq2SeqTrainer, Seq2SeqTrainingArguments, DataCollatorForSeq2Seq

# Load DistilBART model and tokenizer
model_name = "sshleifer/distilbart-cnn-12-6"
bart_tokenizer = BartTokenizer.from_pretrained(model_name)
bart_model = BartForConditionalGeneration.from_pretrained(model_name)
```

```python
def abstractive_summary_bart(text, max_length=100, min_length=30):
    """
    Generates an abstractive summary using the BART model.

    """

    inputs = bart_tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = bart_model.generate(inputs, max_length=max_length, min_length=min_length, length_penalty=2.0, num_beams=4)
    return bart_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Generate and print summary
summary = abstractive_summary_bart(cleaned_text)
print("Abstractive Summary (BART):", summary)
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.text_rank import TextRankSummarizer

def extractive_summary_textrank(text, num_sentences=3):
    """
    Extracts key sentences from the input text using TextRank.
    """
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summarizer = TextRankSummarizer()

    summary = summarizer(parser.document, num_sentences)
    return " ".join(str(sentence) for sentence in summary)

# Generate extractive summary
extractive_summary = extractive_summary_textrank(cleaned_text, num_sentences=2)
print("Extractive Summary (TextRank):", extractive_summary)
```

```python
from datasets import load_dataset

# Load dataset
dataset = load_dataset("cnn_dailymail", "3.0.0")

# Reduce dataset size (use only 5,000 examples for faster training)
small_train_dataset = dataset["train"].select(range(5000))
small_val_dataset = dataset["validation"].select(range(1000))

# Print dataset structure
print(small_train_dataset)
print(small_val_dataset)
```

```python
from datasets import load_dataset

max_input_length = 256  # Reduce from 512 to 256
max_target_length = 32  # Reduce from 64 to 32 (faster summary)

def preprocess_function(examples):
    """
    Tokenizes input articles and corresponding summaries.
    Uses less padding and smaller sequences for speed.
    """
    inputs = bart_tokenizer(
        examples["article"],
        max_length=max_input_length,
        truncation=True,
        padding="max_length"
    )

    targets = bart_tokenizer(
        examples["highlights"],
        max_length=max_target_length,
        truncation=True,
        padding="max_length"
    )

    inputs["labels"] = targets["input_ids"]
    return inputs

# Apply preprocessing
tokenized_train = small_train_dataset.map(
    preprocess_function, batched=True, num_proc=2, remove_columns=["article", "highlights"]
)

tokenized_val = small_val_dataset.map(
    preprocess_function, batched=True, num_proc=2, remove_columns=["article", "highlights"]
)

# Print dataset structure
print(tokenized_train)
print(tokenized_val)
```

```python
from transformers import Seq2SeqTrainingArguments

training_args = Seq2SeqTrainingArguments(
    output_dir="./distilbart_summarization",
    evaluation_strategy="steps",  # Evaluate periodically instead of every epoch
    save_strategy="steps",
    logging_steps=500,  # Reduce logging frequency (less overhead)
    learning_rate=1e-4,  # Increase LR slightly for quicker convergence
    per_device_train_batch_size=1,  # Lower batch size to avoid OOM
    per_device_eval_batch_size=1,
    gradient_accumulation_steps=2,  # Simulate larger batch without OOM
    weight_decay=0.01,
    save_total_limit=1,  # Keep only 1 checkpoint to save space
    num_train_epochs=1,  # Train for just 1 epoch (enough for fine-tuning)
    predict_with_generate=False,  # Disable during training for speed
    logging_dir="./logs",
    save_steps=5000,  # Save less frequently
    fp16=False,  # Keep precision standard (low-end GPUs struggle with fp16)
    gradient_checkpointing=False,  # Disable for speed (uses more memory)
    eval_accumulation_steps=4,  # Optimize evaluation batch processing
)
```

```python
data_collator = DataCollatorForSeq2Seq(bart_tokenizer, model=bart_model)

trainer = Seq2SeqTrainer(
    model=bart_model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    tokenizer=bart_tokenizer,
    data_collator=data_collator
)
```

```python
trainer.train()
```

```python
# Save model and tokenizer
bart_model.save_pretrained("./fine_tuned_bart_summarizer")
bart_tokenizer.save_pretrained("./fine_tuned_bart_summarizer")

print("Model and tokenizer saved successfully!")
```

```python
def summarize_text(text):
    """
    Generates a summary for the given input text.
    """
    inputs = bart_tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = bart_model.generate(inputs, max_length=128, min_length=30, length_penalty=2.0, num_beams=4)
    return bart_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Command-line interaction
if __name__ == "__main__":
    user_input = input("Enter text to summarize: ")
    print("\nSummarized Text:\n", summarize_text(user_input))
```
